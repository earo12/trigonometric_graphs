```python
import plotly.graph_objects as go
import numpy as np
```

```python
# Generar datos
x = np.linspace(0, 2 * np.pi, 100)
y_sin = 3*np.sin(x)
y_cos = np.cos(x)
```

```python
# Crear la figura
fig = go.Figure()

# Agregar la función seno a la figura
fig.add_trace(go.Scatter(x=x, y=y_sin, name='Seno'))
# Agregar la función coseno a la figura
fig.add_trace(go.Scatter(x=x, y=y_cos, name='Coseno'))

# Personalizar el diseño de la gráfica
fig.update_layout(
    title='Gráfica de las funciones Seno y Coseno',
    xaxis_title='X',
    yaxis_title='Y',
    legend_title='Funciones'
)

# Mostrar la gráfica
fig.show()
```
